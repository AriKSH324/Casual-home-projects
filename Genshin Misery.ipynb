```python
start_date = '2020-10-08'
end_date = '2021-03-01'
```

```python
df = pd.read_excel('Genshin_Wish.xlsx', index_col=0)
df3 = df.loc[start_date:end_date].groupby(['Reward Type','Rarity (Star)'])['Wish Count'].count()
```

```python
explode_list = [0.1, 0.2, 0, 0]
df3.plot(kind='pie',figsize=(10,5), autopct='%1.1f%%', shadow = False, explode = explode_list)
plt.title('Genshin Misery Rates')
plt.axis('equal')
```

```python
start_wish = '1'
end_wish = '393'
df4 = df.loc[start_wish:end_wish].groupby(['Rarity (Star)','Reward Type'])['Wish Count'].count()
df4.plot(kind='pie',figsize=(10,5), autopct = '%1.1f%%')
plt.title('Genshin Misery Rates')
plt.axis('equal')
```
