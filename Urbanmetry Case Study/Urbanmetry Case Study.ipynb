# Q2.2

### Plot another price trend that shows the median price psf across time for these three buildings.

```python
Q2_1b = pd.DataFrame(Q2_1.groupby(['QUARTER','NAME'])['psdf_price'].median())
Q2_1b.reset_index(inplace=True)
Q2_1b.rename(columns={"psdf_price":"median"},inplace=True)
Q2_1b['mean'] = Q2_1['psdf_price']
Q2_1b["median"]= Q2_1b["median"].astype(int)
Q2_1b
```

```python
fig, ax = plt.subplots(1,2,figsize = (30,15))

sns.lineplot(data=Q2_1b, x='QUARTER', y='median', hue = 'NAME', ax=ax[0])
sns.lineplot(data=Q2_1b, x='QUARTER', y='mean', hue = 'NAME', ax=ax[1])
```

# Observation
#### Graph data for mean and median has a stark difference, due to mean being average of value and median being the most frequent middle value of a data set sorted from least to greatest. Overall, the psdf_price for each neighbourhood was a steady increase until Q1 2018 where it took a plummet
